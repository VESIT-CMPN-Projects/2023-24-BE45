# **Real data collected by us**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define the data
data = np.array([
    [315, 93],
    [319, 108],
    [317, 130],
    [317, 106],
    [316, 123],
    [311, 121],
    [314, 96],
    [310, 86]
])

# Separate the features (voltage) and target variable (glucose)
X = data[:, 0].reshape(-1, 1)  # Voltage
y = data[:, 1]  # Glucose

# Define models
models = {
    'linr': LinearRegression(),
    'dt': DecisionTreeRegressor(),
    'rf': RandomForestRegressor()
}

# Train and evaluate each model
for name, model in models.items():
    model.fit(X, y)
    predicted_glucose = model.predict(X)
    mae = mean_absolute_error(y, predicted_glucose)
    mse = mean_squared_error(y, predicted_glucose)
    rmse = np.sqrt(mse)
    r2 = r2_score(y, predicted_glucose)
    print(f"Model: {name}")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
    print(f"R-squared (R2): {r2:.2f}")
    print()

    # Plot measured vs predicted glucose levels
    plt.figure(figsize=(8, 6))
    plt.scatter(y, predicted_glucose, color='blue')
    plt.plot([y.min(), y.max()], [y.min(), y.max()], linestyle='--', color='red')
    plt.title(f'{name} - Measured Glucose vs. Predicted Glucose')
    plt.xlabel('Measured Glucose')
    plt.ylabel('Predicted Glucose')
    plt.grid(True)
    plt.show()
```
